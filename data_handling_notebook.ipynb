```python
ds = df.dropna()
```

```python
ds.head()
```

```python
# df['Unnamed: 2'].str.contains('V ugm-3').any()
```

```python
for i in ds.columns:
  if ds[i].str.contains('V ugm-3').any() or ds[i].str.contains('V mgm-3').any():
    ds.drop(columns = i, inplace = True)

ds.head(20)
```

```python
ds.reset_index(inplace= True)
```

```python
ds.head()
```

```python
ds.drop(columns='index', inplace = True)
```

```python
ds.head()
```

```python
# Assuming df is your initial DataFrame

# Create a new DataFrame with the desired column structure
new_columns = []

# First, add the 'Date' column as is
new_columns.append(('Date', ''))

# Iterate through the columns representing stations and pollutants
station_names = []
pollutant_names = []

for i, column_name in enumerate(ds.columns[1:]):
    if i % 4 == 0:
        station_name = column_name
        station_names.extend([station_name] * 4)
    pollutant_names.extend(list(ds.drop(columns='Unnamed: 0').loc[0]))
    # pollutant_names.extend([pollutant_name] * 4)

new_columns.extend(list(zip(station_names, pollutant_names)))

# # Create the new DataFrame with the modified column structure
# new_df = ds.copy()
ds.columns = pd.MultiIndex.from_tuples(new_columns, names=['Station', 'Pollutant'])

# # Now, you have a new DataFrame with the desired multi-index structure
```

```python
ds.head()
```

```python
ds.drop(0, inplace = True)
```

```python
ds.head(3)
```

```python
# Convert the "Date_Column" to the datetime data type
ds['Date'] = pd.to_datetime(ds['Date'], format='%d-%m-%y')

# After running this, the "Date_Column" will be converted to the datetime data type.
```

```python
# Format the date as 'Mon YYYY' and create a new column with the long date format
# ds['Date'] = ds['Date'].dt.strftime('%b %Y')
```

```python
ds.head(3)
```

```python
# Assuming you already have a DataFrame called df
ds.set_index("Date", inplace=True)
```

```python
ds.tail(2)
```

```python
ds.replace("No data", 0, inplace=True)
```

```python
ds.tail(2)
```

```python
ds = ds.astype(float)
```

```python
ds['month'] = ds.index.strftime('%b %Y')
```

```python
ds.tail(2)
```

```python
# Group the DataFrame by the 'month' column and calculate the mean for each group
final = ds.groupby(by=["month"]).mean()
```

```python
# Convert the index to datetime format
final.index = pd.to_datetime(final.index, format='%b %Y')

# Sort the DataFrame based on the datetime index
k = final.sort_index()
```

```python
k
```

```python
k.index = k.index.strftime('%b %Y')
```
