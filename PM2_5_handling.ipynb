```python
ds = ds.replace('No data', np.nan)
```

```python
ds = ds.astype(float)
```

```python
ds['month'] = ds.index.strftime('%b %Y')
```

```python
ds.tail(2)
```

```python
# Group the DataFrame by the 'month' column and calculate the mean for each group
final = ds.groupby(by=["month"]).mean()
```

```python
# Convert the index to datetime format
final.index = pd.to_datetime(final.index, format='%b %Y')

# Sort the DataFrame based on the datetime index
k = final.sort_index()
```

```python
k.index = k.index.strftime('%b %Y')
```

```python
# Assuming your DataFrame is named 'df' and you want to save it to 'output.csv'
# k.T.to_csv('output.csv')  # Use index=False to exclude row indices in the output
```

```python
final = k.T
```

```python
final.head()
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import pandas as pd

# Assuming 'M' is your original DataFrame
# Create an empty DataFrame 'X' with station names as row indices
X = pd.DataFrame(index=final.index.levels[0][1:-1])

# # Iterate through rows in 'M' and populate 'X'
for month in final.columns:
  for station, pollutant in final.index:
    column_name = f"{month}-{pollutant}"
    X.loc[station, column_name] = final.loc[(station, pollutant), month]
X
```

```python
Average_data = X.copy()
```

```python
# List of column names containing 'Carbon monoxide'
columns_to_multiply = [col for col in Average_data.columns if 'Carbon monoxide' in col]

# Multiply the selected columns by 1000
Average_data[columns_to_multiply] = Average_data[columns_to_multiply] * 1000
```
